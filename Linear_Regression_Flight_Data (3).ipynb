The `add_constant` method from the `statsmodels` library is used to add an intercept term to your model. When you're performing linear regression, including an intercept is a standard approach unless you have specific reasons to omit it. The intercept represents the expected mean value of Y when all X variables (x1, x2, x3 in your case) are set to 0.

In other words, it gives the baseline value of your dependent variable (Y) when all independent variables (X) are zero. It's crucial for ensuring that your linear model has a y-intercept which is more realistic for most real-world scenarios. Without this constant, your linear model would be forced to go through the origin (0,0), which often doesn't make sense in practice, especially if there's no reason to believe that Y should be zero when all Xs are zero.

Including a constant in the model also improves the interpretation of the regression coefficients. Each coefficient represents the change in Y for a one-unit change in the respective X variable, holding other variables constant, and this interpretation assumes that there is a baseline Y value (the intercept) when all Xs are zero.

So, by using `sm.add_constant(X)`, you ensure that your linear regression model correctly estimates and includes this intercept term. This makes your model's predictions and inferences more accurate and interpretable.

The output you've provided is from an OLS (Ordinary Least Squares) regression model summary. Each term represents different aspects of the model and its performance:

### General Terms:

- **Dep. Variable (Dependent Variable)**: This is the variable that the model is trying to predict, in this case, 'Passengers'.
- **Model**: Indicates the type of model used; here it's OLS, meaning Ordinary Least Squares, a standard method for estimating the linear regression coefficients.
- **Method**: Describes the method used to fit the data, which is 'Least Squares'. This means the model minimizes the sum of the squares of the differences between observed and predicted values.
- **No. Observations**: The number of observations (data points) used in the model, which is 80 in this case.
- **Df Residuals**: Degrees of freedom of the residuals; calculated as the number of observations minus the number of parameters being estimated. Here, it is 79.
- **Df Model**: Number of predictors used in the model, here it's 1 (usually this doesn't include the intercept).

### Fit Quality Metrics:

- **R-squared (uncentered)**: The proportion of the variance in the dependent variable that is predictable from the independent variable(s), which is 0.996 here. However, it's 'uncentered' because it doesn't compare against the mean of the dependent, which is typical in models without a constant term.
- **Adj. R-squared (uncentered)**: Adjusted R-squared considers the number of predictors in the model and adjusts the R-squared value accordingly; here it is also 0.996.
- **F-statistic**: A measure to test if there is a significant relationship between the dependent variable and the independent variable(s). A higher value indicates a stronger relationship.
- **Prob (F-statistic)**: The p-value corresponding to the F-statistic, telling us the probability that the null hypothesis for the full model is true. Here, it's extremely low (1.70e-96), suggesting the model is statistically significant.

### Model Coefficients:

- **coef (Coefficient)**: Indicates the change in the dependent variable for a one-unit change in the predictor. For 'Promotion_Budget', a one-unit increase corresponds to a 0.0714 increase in 'Passengers'.
- **std err (Standard Error)**: The standard error of the coefficient, giving a measure of the variability in the estimate for the coefficient.
- **t**: The t-statistic value for testing the null hypothesis that the coefficient equals zero (no effect).
- **P>|t|**: The p-value associated with the above t-statistic. A low value (<0.05) typically indicates that the predictor is a significant variable in the model.
- **[0.025 0.975]**: The 95% confidence interval for the coefficient; in this case, it's between 0.070 and 0.072.

### Diagnostics:

- **Log-Likelihood**: The log-likelihood of the model, used in more complex model selection criteria.
- **AIC (Akaike Information Criterion)**: A measure for model comparison; lower values indicate a better model, given a balance of goodness of fit and complexity.
- **BIC (Bayesian Information Criterion)**: Similar to AIC but with a higher penalty for models with more parameters, thus it's also used for model comparison.
- **Covariance Type**: Indicates the type of covariance matrix (here 'nonrobust'), which is an assumption about the error terms in the model.
- **Durbin-Watson**: Tests for autocorrelation in the residuals from a regression; values near 2 indicate no autocorrelation.
- **Omnibus, Prob(Omnibus)**: Tests the hypothesis that the residuals are normally distributed; a lower value suggests deviation from normality.
- **Jarque-Bera (JB), Prob(JB)**: Another test for normality in the residuals. A lower value here suggests non-normality.
- **Skew**: Measures the asymmetry of the residual distribution. Near zero suggests symmetry.
- **Kurtosis**: Measures the 'tailedness' of the residual distribution; further from 3 indicates departure from normal distribution.
- **Cond. No. (Condition Number)**: Assesses the multicollinearity (when independent variables are highly correlated); high values indicate potential numerical problems in the regression.

In summary, this model summary provides information on how well the model fits, the significance of the model coefficients, and diagnostic information about the model residuals.

In the context of linear regression, an ANOVA table serves a slightly different purpose but follows a similar structure. Let me explain:

In linear regression, the ANOVA table is used to assess the overall fit of the model and test the significance of the regression. Here's how it's structured:

1. Sources of Variation:
   - Regression (or Model): Variation explained by the regression model
   - Residual (or Error): Unexplained variation
   - Total: Total variation in the dependent variable

2. Degrees of Freedom (df):
   - Regression: Number of predictors (k)
   - Residual: Sample size minus number of predictors minus 1 (n - k - 1)
   - Total: Sample size minus 1 (n - 1)

3. Sum of Squares (SS):
   - Regression: Explained sum of squares
   - Residual: Unexplained sum of squares
   - Total: Total sum of squares

4. Mean Square (MS):
   - Calculated by dividing SS by respective df

5. F-statistic:
   - Ratio of Mean Square Regression to Mean Square Residual

6. p-value:
   - Probability of obtaining the observed F-statistic under the null hypothesis

The ANOVA table in regression helps determine:
1. Overall model fit (Is the model statistically significant?)
2. Proportion of variance explained (R-squared can be calculated from SS)
3. Significance of the regression relationship

#Heteroscedasticity

Heteroscedasticity is an important concept in linear regression that refers to the unequal variance of residuals across the range of measured values. Let me break this down:

1. Definition:
   - In a regression model, heteroscedasticity occurs when the variability of the residuals (the differences between predicted and actual values) is not constant across all levels of the independent variable(s).
   - The opposite of heteroscedasticity is homoscedasticity, where residuals have constant variance.

2. In the context of linear regression:
   - One of the key assumptions of ordinary least squares (OLS) regression is that the residuals have constant variance (homoscedasticity).
   - When this assumption is violated, we have heteroscedasticity.

3. Causes:
   - Outliers in the data
   - Omitted variables from the model
   - Incorrect functional form of the model
   - Measurement errors

4. Consequences:
   - While heteroscedasticity doesn't bias the coefficient estimates, it can lead to:
     - Inefficient estimates (not minimum variance)
     - Biased standard errors
     - Incorrect inference (t-tests and F-tests become unreliable)

5. Detection:
   - Visual methods: Plotting residuals vs. fitted values or independent variables
   - Statistical tests: Breusch-Pagan test, White test, Goldfeld-Quandt test

6. Remedies:
   - Weighted least squares regression
   - Transforming variables (e.g., log transformation)
   - Using robust standard errors
   - Generalized least squares (GLS) estimation

7. Importance:
   - Addressing heteroscedasticity is crucial for valid statistical inference and accurate predictions, especially when dealing with cross-sectional data or time series with varying volatility.

Understanding and addressing heteroscedasticity is essential for ensuring the reliability and validity of linear regression models, particularly in fields like economics, finance, and social sciences where data often exhibits this characteristic.

````python
# prompt: give the code to calculate Heteroscedasticity for the above regression

import statsmodels.stats.api as sms
import matplotlib.pyplot as plt

# Breusch-Pagan test for heteroscedasticity
bp_test = sms.het_breuschpagan(fitted1.resid, fitted1.model.exog)

# Print the test results
print('Breusch-Pagan test for heteroscedasticity:')
print(f'Test statistic: {bp_test[0]:.3f}')
print(f'p-value: {bp_test[1]:.3f}')

# White test for heteroscedasticity
white_test = sms.het_white(fitted1.resid, fitted1.model.exog)

# Print the test results
print('\nWhite test for heteroscedasticity:')
print(f'Test statistic: {white_test[0]:.3f}')
print(f'p-value: {white_test[1]:.3f}')

# Plot residuals versus fitted values
plt.scatter(fitted1.fittedvalues, fitted1.resid)
plt.axhline(0, color='red')
plt.xlabel('Fitted values')
plt.ylabel('Residuals')
plt.title('Residuals vs. Fitted values')
plt.show()
````

# L1 Regularization (LASSO)

## LASSO (Least Absolute Shrinkage and Selection Operator) is a type of linear regression that adds a penalty term to the loss function. This penalty term is proportional to the absolute value of the coefficients, which encourages the model to select a subset of relevant features and set the coefficients of irrelevant features to zero.

## This can be useful in cases where there are many features and only a few are truly relevant, as it can help to prevent overfitting and improve model interpretability.

## The hyperparameter alpha controls the amount of regularization. A higher value of alpha leads to more features being set to zero, while a lower value of alpha leads to less regularization.

## The following code demonstrates how to fit a LASSO model using scikit-learn:

lasso_model = Lasso(alpha=0.1)
lasso_model.fit(X, y)

# Print the coefficients
print('Lasso coefficients:', lasso_model.coef_)

# L2 Regularization (Ridge)

## Ridge regression is another type of linear regression that adds a penalty term to the loss function. This penalty term is proportional to the squared value of the coefficients, which encourages the model to have smaller coefficients but does not set any coefficients to zero.

## This can be useful in cases where there are many features and all of them are likely to be relevant, but we want to prevent overfitting.

## The hyperparameter alpha controls the amount of regularization. A higher value of alpha leads to smaller coefficients, while a lower value of alpha leads to less regularization.

## The following code demonstrates how to fit a Ridge model using scikit-learn:

ridge_model = Ridge(alpha=0.1)
ridge_model.fit(X, y)

## Print the coefficients
print('Ridge coefficients:', ridge_model.coef_)


````python
# prompt: give the code for L1 & L2 Regularization for above regression

# L1 Regularization (LASSO)
from sklearn.linear_model import Lasso

lasso_model = Lasso(alpha=0.1)
lasso_model.fit(X, y)

# Print the coefficients
print('Lasso coefficients:', lasso_model.coef_)

# L2 Regularization (Ridge)
from sklearn.linear_model import Ridge

ridge_model = Ridge(alpha=0.1)
ridge_model.fit(X, y)

# Print the coefficients
print('Ridge coefficients:', ridge_model.coef_)
````
